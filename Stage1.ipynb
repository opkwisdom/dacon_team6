🙋 `pandas` 라이브러리가 무엇인가요?<br>
판다스 라이브러리는 데이터 분석을 위한 라이브러리로, Python 프로그래밍 언어를 위한 라이브러리입니다!<br>
빠르고 범용성이 좋기 때문에 데이터 분석에 필수적인 라이브러리라고 할 수 있죠!<br>
판다스 라이브러리는 이번에 저희가 다루게 될 데이터처럼 **행(row)**과 **열(column)**로<br> 이루어진 2차원 데이터를 다루는데 유용한 패키지입니다.

🙋 그렇다면 `CSV` 파일은 무엇인가요?<br>
CSV 파일은 **Comma-Separated Values**의 약자로, 몇 가지 필드를 **쉼표(,)**로 구분한 텍스트 파일을 말합니다.<br>
데이터 분석 프로젝트의 일반적인 데이터 형식으로, CSV 파일은 매우 빈번하게 사용된답니다!

* 📝 `test.csv`: 학습된 모델을 바탕으로 결과를 예측하기 위한 데이터 파일입니다!
* 🚩 `sample_submission.csv`: `test.csv`을 통해 예측된 결과를 제출하기 위한 파일입니다!

데이터프레임에는 `head()`와 `tail()` 메서드가 존재합니다!<br>
`head()` 메서드는 데이터프레임(DataFrame)의 첫 번째부터 지정된 개수의 행(row)을 반환하는 메서드입니다.<br>
반면, `tail()` 메서드는 데이터프레임(DataFrame)의 마지막부터 지정된 개수의 행(row)을 반환하는 메서드입니다.<br>
`head()`와 `tail()` 메서드는 일부 행을 확인하고 데이터의 구조와 내용을 파악하는 데 사용되는데,<br>
특히 큰 데이터프레임이나 데이터셋에서 데이터의 형태를 파악하고자 할 때 주로 사용합니다.

```python
submission.head()
```
