### **2. describe()**

두 번째로 알아볼 메서드는 `describe()` 메서드입니다.<br>
`describe()` 메서드는 데이터프레임의 기초 통계량을 나타내주는 메서드입니다!<br>
`describe()` 메서드는 아래와 같은 정보를 표시해줍니다.

📜 **기초 통계량**<br>
* count: 데이터 개수
* mean: 평균
* std: 표준편차
* min: 최솟값
* 25%: 제1 사분위수 (Q1)
* 50%: 제2 사분위수, 중간값 (Q2)
* 75%: 제3 사분위수 (Q3)
* max: 제4 사분위수 (Q4)
* unique (수치형 X): 서로 다른 값의 개수
* top (수치형 X): 최빈값
* freq (수치형 X): 최빈값의 빈도

`describe()`는 주로 수치형 데이터를 간략하게 파악하고 싶을 때 유용하게 사용할 수 있습니다.<br>
